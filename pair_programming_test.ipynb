```python
import numpy as np

def polar_to_cartesian(r, theta):
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return x, y
    
```

```python
# Test 1: theta = 0 should result in a point on the x-axis

x, y = polar_to_cartesian(4, 0)

print(x, y)
```

```python
# Test 2: theta = pi/2 should result in a point on the y-axis

x, y = polar_to_cartesian(5, np.pi/2)

print(x, y)
```
