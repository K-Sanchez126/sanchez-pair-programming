```python
import numpy as np
#Function that takes r and theta as input and outputs the x and y cooridnates 
def polar_to_cartesian(r, theta):
    x= r * np.cos(theta)
    y = r* np.sin(theta)
    return x,y
    
```

```python
# Test 1: angle 0 should give a point on the positive x-axis

x, y = polar_to_cartesian(4, 0)

print("Test 1 output:", x, y)

assert np.isclose(x, 4)
assert np.isclose(y, 0)

print("Test 1 passed")
```

```python
# Test 2: angle pi/2 should give a point on the positive y-axis

x, y = polar_to_cartesian(5, np.pi / 2)

print("Test 2 output:", x, y)

assert np.isclose(x, 0)
assert np.isclose(y, 5)

print("Test 2 passed")
```
